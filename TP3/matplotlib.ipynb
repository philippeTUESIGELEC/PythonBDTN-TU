    matplotlib.pyplot est une collection de fonctions de commande qui fait marcher matplotlib comme MATLAB. Chaque fonction pyplot réalise des changements sur une figure. Exemple : créer une figure, créer un graphique, dessiner des lignes, décorer des graphiques avec des labels. Dans matplotlib.pyplot, différents états sont préservés à travers les appels, afin qu'il garde une trace des choses comme la figure actuelle et la zone de tracé, les fonctions de graphiques sont dirigés à l'axe actuel (notez que "axes" ici se réfère la plupart du temps aux axes de la figure et pas dans un sens strict mathématiques) 

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot([1,2,3,4])
plt.ylabel('des nombres')
plt.show()
```

Vous vous demandez peut-être pourquoi l'axe x va de 0 à 3 et l'axe Y va de 1 à 4. Si vous fournissez une seule liste ou tableau à la commande plot(). Matplotlib considère cela comme une séquence de valeur Y, et automatiquement génère des valeurs x pour vous. Du fait que la séquence Python commence à 0, le vecteur par défaut X a la meme longueur que Y mais commence à 0. Par conséquent, les données X sont [0,1,2,3]
plot() est une commande versatile, et va prendre un nombre arbitraire d'arguments. Par exemple, pour tracer un graphique X contre Y, vous pouvez utiliser la commande suivante.

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

Pour chaque pair de x,y, il y a un troisième argument optionnel qui est le format caractère qui indique le type de couleur et de ligne du tracé. Les lettres et les symboles du format chaine sont de MATLAB, et vous pouvez concatener une couleur chaine avec un chaine de style ligne. Le format de chaine par défaut est 'b-, ce qui est une ligne bleue solide. Par exemple, pour tracer le cercle rouge suivant, vous utiliseriez.

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.show()
```

Regardez la documentation plot() pour une liste complète de style de liste de ligne et des format de chaines de caractères. La commande axis() dans l'exmple ci-dessus prend une liste de [xmin,xmax,ymin,ymax] et spécifie la vue des axes.
Si matplotlib était limité à ses listes, il serait inutile pour processer des valeurs numériques. Généralement, vous utiliserez des tableaux numpy. En fait, toutes les séquences sont converties en tableaux numpy en interne. L'exemple ci-dessous illustre un tracé de plusieurs lignes avec différents styles de format en une commande utilisant des tableaux.

```python
import numpy as np
import matplotlib.pyplot as plt
#Echantillone équitablement le temps à 200 ms d'intervalle
t = np.arange(0., 5., 0.2)
#Tirets rouges, carres bleus et triangles verts
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

#Travailler avec plusieurs figures et axes
MATLAB et pyplot, ont le concept de figure actuelle et d'axes actuels. Toute les commandes de tracés s'appliquent aux axes actuelles. La fonction gca() retourne les axes actuelles ( instancematplotlib.axes.Axes), et gcf() retourne la figure actuelle.
(instance matplolib.figure.Figure). Normalement, vous n'avez pas à vous inquiéter pour ça, parce tout est géré derrière. Ci-dessous un script pour créer deux sous graphiques.

```python
import numpy as pd
import matplotlib.pyplot as plt
%matplotlib inline
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

La commande figure() ici est optionnel parce figure(1) sera créé par défaut, tout comme subplot(111) sera créé par défaut si vous ne spécifiez pas d'axes manuellement. La commande subplot() spécifie numrows,numcols,fignum ou fignum va de 1 à numrows*numrows. Les virgules dans la commande subplot sont optionnels si numrows*numcols<10. Alors subplot(211) est identique à subplot(2,1,1). Vous pouvez creer un nombre arbitraire de sous graphiques et des axes. Si vous voulez placer des axes manuellement, pas sur une grille rectangulaire, utilisez la commande axes(), qui vous permet de spécifier le lieu tel que axes([left,bottom,width,height]) ou toute les valeurs sont fractionnel (0 à 1).
Vous pouvez créer plusieurs graphiques en utilisant plusieurs appels figure() avec un nombre croissant de graphiques. Bien sur, chaque figure peut contenir autant d'axe et de sous graphiques.

```python
import matplotlib.pyplot as plt
plt.figure(1)                # premiere figure
plt.subplot(211)             # le premier sous graphique dans le premier graphique
plt.plot([1, 2, 3])
plt.subplot(212)             # le second sous graphique dans le premier graphique
plt.plot([4, 5, 6])


plt.figure(2)                # une seconde figure
plt.plot([4, 5, 6])          # creer un sous graphique subplot(111) par défaut

plt.figure(1)                # figure 1 actuel; subplot(212) encore actuelle
plt.subplot(211)             # fait subplot(211) une figure actuelle 
plt.title('Easy as 1, 2, 3') # titre de subplot 211
```

Vous pouvez nettoyer la figure actuelle avec clf() et les axes actuels avec cla(). Si vous trouvez cela énervant, cet appel (cet image, figure et axes) sont maintenus pour vous en interne, ne désespérez pas.
Si vous faites beaucoup de graphiques, vous devez savoir une chose: la mémoire requise pour un graphique n'est pas complètement faites jusqu'à que la figure est explicitement close(). Supprimer toutes les références de la figure et/ou utiliser le gestionnaire de fenêtre pour fermer la fenetre dans lequel la figure apparait sur l'écran.

#Travailler avec du texte
La commande text() peut être utilisé pour ajouter du texte dans un lieu arbitraire, et xlabel(),ylabel() et title() sont utilisés pour ajouter du texte dans un lieu indiqué.

```python
import numpy as np
import matplotlib.pyplot as plt

#Fixer des etats aléatoires pour la reproductibilité
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

#Histogramme de données
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

Toute la command text() retourne une instance matplotlib.text.Text. Just avec les lignes au dessus, vous pouvez personnaliser les propriétés en passant les mots clés en arguments dans la fonction text ou en utilisant setp() : 

```python
t = plt.xlabel('my data', fontsize=14, color='red')
```

Ces propriétés sont converts dans les détails dans Les propriétés textes et couches.

Utiliser les expressions mathématiques dans le texte.
Matplotlib accepte les équations TeX dans chaque expression text. Il faut utiliser des signes dollars

#Texte d'annotation
L'utilisation de la commande text() au dessus place du texte à une position arbitraire sur les axes. Une utilisation commune pour du texte est d'annoter des fonctionnalités du tracé, et la méthode annotate() fournit des fonctionnalités d'aide pour rendre les annotations faciles. Dans une annotation, il y a deux points à considerer : le lieu qui est annoté représenté par l'argument xy et le lieu du texte xytext. Deux de ses arguments sont des tuples. 

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.ylim(-2,2)
plt.show()
```

Dans cet exemple basique, les lieux de xy (arrow tip) et xytext (lieu de texte) sont dans des données de coordonnées. Il existe une autre variété de systèmes de coordonnées.

#Logarithme et autres axes nonce linéaires
matplotlib.pyplot supporte les échelles d'axes non linéaires, mais aussi les échelles logarithmiques et logit. C'est communément utilisé si la donnée s'étend sur des ordres de magnitude. Changer l'échelel d'un axe est simple : 
plt.xscqle('log')
Un exemple de quatre graphques des memes données et différentes échelles pour l'axe y est montré ci-dessous.

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter  # utile pour l'échelle logit

# Fixer des états aléatoires pour la reproductibilité
np.random.seed(19680801)

# maquiller des données dans l'intervalle ]0,1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# trqcer qvec différentes axes d'échelles
plt.figure(1)

# linéaire
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# log symétrique
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Formate les labels de l'axe y dans des chaines vides 
# 'NullFormatter', pour éviter d'emcombrer l'axe avec trop de label
plt.gca().yaxis.set_minor_formatter(NullFormatter())
#Ajuster la couche du sous graphique, le logit peut prendre de l'espace.
# Comme d'habitude, du au label y
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.show()
```
